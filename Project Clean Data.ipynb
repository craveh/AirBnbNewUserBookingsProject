```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
train_users = pd.read_csv('airbnb-recruiting-new-user-bookings/train_users_2.csv')
train_users.info()
train_users.head()
```

```python
demo = pd.read_csv('airbnb-recruiting-new-user-bookings/age_gender_bkts.csv')
demo.info()
demo.head()
# demo['year'].unique()
```

```python
countries = pd.read_csv('airbnb-recruiting-new-user-bookings/countries.csv')
countries.info()
countries.head()
```

```python
sessions = pd.read_csv('airbnb-recruiting-new-user-bookings/sessions.csv')
sessions.info()
sessions.head()
```

```python
test_users = pd.read_csv('airbnb-recruiting-new-user-bookings/test_users.csv')
test_users.info()
test_users.head()
```

```python
test_ = pd.merge(left= train_users, right=sessions, left_on='id', right_on='user_id')
```

```python
test_.info(verbose=True, show_counts=True)
# test_.head(15)
# test_.tail(20)
```

```python
with_drop = test_.dropna()
```

```python
with_drop.info(verbose=True, show_counts=True)
with_drop.head()
```

```python
clean_unknown = with_drop.replace(['-unknown-'], np.nan)
clean_unknown.info(show_counts=True)
```

```python
unknown_dropped_df = clean_unknown.dropna()
unknown_dropped_df.info(show_counts=True)
```

```python
# unknown_dropped_df.sample(5)
unknown_dropped_df.head()
```
